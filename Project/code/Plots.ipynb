```python
precision_data = pd.read_csv('../results/correct_results/results_for_stats_precision.csv', header=None)
recall_data = pd.read_csv('../results/correct_results/resutls_for_stats_recall.csv', header=None)
```

```python
precision_data = precision_data.transpose()
recall_data = recall_data.transpose()
columns = ['NB', 'DRONE (default)', 'DRONE (v2)', 'RF_notune', 'RF_19-13-1', 'RF_29-10-1', 'RF_46-13-1']
precision_data.columns = columns
recall_data.columns = columns
```

```python
precision_data.to_csv('../results/correct_results/results_for_stats_precision_df.csv', index=False)
recall_data.to_csv('../results/correct_results/results_for_stats_recall_df.csv', index=False)
```

```python
columns = columns[::-1]
data_plot_precision = []
for c in columns:
    data_plot_precision.append(list(precision_data[c]))
    
data_plot_recall = []
for c in columns:
    data_plot_recall.append(list(recall_data[c]))
```

```python
colors = ['Green', 'Black', 'Green', 'Blue', 'Blue', 'Red', 'Red']
fig = plt.figure(1, figsize=(20, 10))

ax = fig.add_subplot(111)
ax.grid()
# Create the boxplot
bp = ax.boxplot(data_plot_recall)



i = 0
for box in bp['boxes']:
    box.set(color=colors[i])
    i+=1
#     box.set(face_color='DarkGreen')

plt.xticks(range(1, len(columns)+1),columns)
plt.ylabel('Recall Score')
plt.title('Model Comparison')
plt.show()
```
